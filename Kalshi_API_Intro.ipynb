```python
import requests_cache
from tqdm import tqdm

session = requests_cache.CachedSession('/content/drive/MyDrive/POLYMARKET/requests_cache')

cursor = ""
total_markets = float('inf')  # We'll update this later
events = []
with tqdm(total=total_markets, desc="Fetching markets") as pbar:
  while True:
    url = "https://trading-api.kalshi.com/trade-api/v2/events"
    headers = {"accept": "application/json"}

    response = session.get(url, headers=headers,params=dict(
        limit=200,
        cursor=cursor,
        with_nested_markets=True
    ))
    r=response.json()
    if r.get("cursor")==cursor or not r.get("cursor"):
      break
    events.extend(r.get("events"))

    cursor=r.get("cursor")
    pbar.update(len(events))
```

```python
import asyncio
import websockets
import json

KALSHI_WEBSOCKET = "wss://trading-api.kalshi.com/trade-api/ws/v2"

headers={"Authorization": f"Bearer {response['token']}"}

trade_message = '{"id": 1, "cmd": "subscribe", "params": {"channels": ["trade"]}}'
# fill_message = '{"id": 2, "cmd": "subscribe", "params": { "channels": ["fill"]}}'
#orderbook_delta_message = '{"id": 3, "type": "subscribed", "msg": {"channel": "orderbook_delta", "sid": 1}}'

async def process_message(message):
    print(f"Received message: {message}")


async def listen_to_server():
    async with websockets.connect(KALSHI_WEBSOCKET, extra_headers=headers) as websocket:
        await websocket.send(trade_message)
        print(f"Sent trade message: {trade_message}")

        while True:
            try:
                message = await websocket.recv()
                await process_message(message)
            except websockets.ConnectionClosed:
                print("Connection closed")
                break

loop = asyncio.get_running_loop()
await loop.create_task(listen_to_server())
```
